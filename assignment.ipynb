# **Convert timestamps and align the datasets by date (daily level is fine)
**

```python

fear_df["date"] = pd.to_datetime(fear_df["date"])


history_df["Timestamp IST"] = pd.to_datetime(
    history_df["Timestamp IST"],
    format="%d-%m-%Y %H:%M"
)


fear_df["date"] = fear_df["date"].dt.date
history_df["date"] = history_df["Timestamp IST"].dt.date


merged_df = pd.merge(
    history_df,
    fear_df,
    on="date",
    how="left"
)


merged_df.head()
```

# **Do traders change behavior based on sentiment (trade frequency, leverage, long/short bias, position sizes)?

bold text**

```python
trade_count = merged_df.groupby("Account").size().reset_index(name = "Trade Count")


threshold = trade_count["Trade Count"].quantile(0.75)

trade_count["Trader Type"] = np.where(
    trade_count["Trade Count"] >= threshold,
    "Frequent",
    "Infrequent"
)
```

```python
merged_df = merged_df.merge(
    trade_count[["Account", "Trader Type"]],
    on="Account",
    how="left"
)
```

```python
avg_profit = (
    merged_df.groupby("Trader Type")["Closed PnL"]
    .mean()
    .reset_index(name="Average Profit")
)

avg_profit
```

```python
avg_position_size = (
    merged_df.groupby("Trader Type")["Size USD"]
    .mean()
    .reset_index(name="Average Position Size")
)

avg_position_size
```

```python
merged_df["Win"] = merged_df["Closed PnL"] > 0
win_rate = (
    merged_df.groupby("Trader Type")["Win"]
    .mean() * 100
)

win_rate
```

```python
merged_df["Win"] = merged_df["Closed PnL"] > 0
win_rate = (
    merged_df.groupby("Account")["Win"]
    .mean()
    .reset_index(name="Win Rate")
)

win_rate["Win Rate"] = win_rate["Win Rate"] * 100
```

```python
win_rate["Trader Segment"] = np.where(
    win_rate["Win Rate"] >= 60,
    "Consistent Winner",
    "Inconsistent Trader"
)
```

```python
merged_df = merged_df.merge(
    win_rate[["Account", "Trader Segment"]],
    on="Account",
    how="left"
)
```

# **Provide at least 3 insights backed by charts/tables.
**

```python
avg_pnl = merged_df.groupby("classification")["Closed PnL"].mean()
win_rate = merged_df.groupby("classification")["Win"].mean() * 100

performance = pd.DataFrame({
    "Average PnL": avg_pnl,
    "Win Rate (%)": win_rate
})

performance
```

```python
performance["Average PnL"].plot(kind="bar", figsize=(6,4))

plt.title("Average PnL by Market Sentiment")
plt.xlabel("Market Sentiment")
plt.ylabel("Average PnL")

plt.xticks(rotation=0)
plt.show()
```

```python
performance["Win Rate (%)"].plot(kind="bar", figsize=(6,4))

plt.title("Win Rate by Market Sentiment")
plt.xlabel("Market Sentiment")
plt.ylabel("Win Rate (%)")

plt.xticks(rotation=0)
plt.show()

```
